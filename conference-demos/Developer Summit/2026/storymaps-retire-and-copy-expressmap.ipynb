## 🎯 User Story

We have a multi-year-old story covering a project that is now complete. We will take this story offline.

However, this story has a complex **Express map** or **Sidecar** that we would like to keep and use in other combined project's storymap, where it can be used in a new context.

````python
import arcgis
from arcgis.gis import GIS
from arcgis.apps.storymap import StoryMap
from arcgis.apps.storymap.story_content import ExpressMap, Map
from dotenv import load_dotenv
import shutil
import os
````

````python
arcgis.__version__ = '2.4.3'
````

````python
## load environment variables from .env file
load_dotenv()
## get credentials from environment variables
USERNAME_TARGET = os.getenv("USERNAME_TARGET")
PASSWORD_TARGET = os.getenv("PASSWORD_TARGET")
````

````python
#login to arcgis
gis_con = GIS("https://www.arcgis.com", USERNAME_TARGET, PASSWORD_TARGET)
````

````python
## CONSTANTS
SOURCE_ID = "" # Source StoryMap item ID
TARGET_ID = "" # Target StoryMap item ID (can be the same as source if you want to overwrite it)
````

### Define source and tagets

````python
source_storymap = StoryMap(SOURCE_ID, gis_con)
target_storymap = StoryMap(TARGET_ID, gis_con)
````

## Find the expressmap
To find the expressmap we will walk through the source storymaps nodes/blocks to find the expressmap node. Achieved by looking for node type and alt text combination.

```json
//...properites
    "nodes":{
        // ...
    "n-Ala5eE":{
        "type": "expressmap",
        "data": {
            "map": "r-eiFx65",
            "basemap": "modernAntique",
            "alt": "Target"
        }
    }
    //..nodes continued
}
```

````python
# get node of expressmap with alt text
target_expressmap = None
for node_id, node in source_storymap._properties["nodes"].items():
    if node["type"] == "expressmap" and node['data'].get("alt") == "Target":
        print(f"Found expressmap with alt text 'Target': {node_id}")
        target_expressmap = node_id
````

## Create Express Map Instance

Loop through the source storymap's content list to select the express map with the matching node_id.

````python
# Grab the Expressmap
expressmap = None
for content_item in source_storymap.contents:
    if isinstance(content_item, ExpressMap) and content_item._node == target_expressmap:
        print(content_item.node)
        print(f"Found ExpressMap content item with node {target_expressmap}")
        expressmap = content_item
````

## Copy that expressmap

````python
source_storymap.copy_content(target_story=target_storymap, node_list=[expressmap])
````
